```python
class LinearRegressionGD(object):
    def __init__(self, eta=0.001, n_iter=20):
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self, X, y):
        self.w_ = np.zeros(1 + X.shape[1])
        self.w_ = self.w_.reshape(w_.shape[0],1)
        self.cost_ = []
        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (y - output)
            
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self
    
    def net_input(self,X):
        return np.dot(X,self.w_[1:] + self.w_[0])
    
    def predict(self,X):
        return self.net_input(X)
```

```python
X = df[['RM']].values
y = df[['MEDV']].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
sc_y = StandardScaler()
```

```python
X_std = sc_x.fit_transform(X)
```

```python
y_std = sc_y.fit_transform(y)
```

```python
lr = LinearRegressionGD()
```

```python
lr.fit(X_std, y_std)
```

```python
plt.plot(range(1,lr.n_iter+1), lr.cost_)
plt.ylabel('Cost')
plt.xlabel('Number of iterations')
```

```python
def lin_regplot(X,y,model):
    plt.scatter(X,y,c='blue')
    plt.plot(X,model.predict(X),color='red')
    return None
```

```python
lin_regplot(X_std, y_std, lr)
plt.xlabel('Avg no. of rooms [RM] (standardized)')
plt.ylabel('Housing price [thousands of $] (standardized)')
```

```python
num_rooms_std = sc_x.transform([5.0])
```

```python
print 'Price in $1K: %.3f' % sc_y.inverse_transform(price_std)
```

## Random sample consensus (RANSAC)

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor(LinearRegression(), 
                         max_trials=100, 
                         min_samples=50,
                         residual_metric=lambda x:np.sum(np.abs(x), axis=1),
                         residual_threshold=5.0,
                         random_state=0)
```

```python
ransac.fit(X,y)
```

```python
inlier_mask = ransac.inlier_mask_
```

```python
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3,10,1)
```

```python
line_y_ransac = ransac.predict(line_X[:, np.newaxis])
```

```python
plt.scatter(X[inlier_mask], y[inlier_mask], 
            c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask], 
            c='lightgreen', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac,color='red' )
plt.xlabel('Avg number of rooms [RM]')
plt.ylabel('Price in $1K [MEDV]')
plt.legend(loc='upper left')
```

```python
print('Slope: %.3f' % ransac.estimator_.coef_[0])
```

```python
print('Intercept: %.3f' % ransac.estimator_.intercept_)
```

## Evaluate performance of linear regression model

```python
from sklearn.cross_validation import train_test_split
```

```python
X = df.iloc[:,:-1].values
y = df['MEDV'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.3,random_state=0)
```

```python
slr = LinearRegression()
```

```python
slr.fit(X_train, y_train)
```

```python
y_train_pred = slr.predict(X_train)
```

```python
y_test_pred = slr.predict(X_test)
```

```python
plt.scatter(y_train_pred, y_train_pred-y_train,
            c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred-y_test, 
            c='lightgreen', marker='s', label='Test data')

plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='red')
plt.xlim([-10,50])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print 'MSE train: %.3f, test: %.3f' % \
(mean_squared_error(y_train, y_train_pred), \
 mean_squared_error(y_test, y_test_pred))
```

```python
from sklearn.metrics import r2_score
```

```python
print 'R^2 train: %.3f, test: %.3f' % \
(r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred))
```

## Model nonlinear relationships in housing data

```python
X = df[['LSTAT']].values
y = df[['MEDV']].values
```

```python
regr = LinearRegression()
```

```python
# create polynomial features
quadratic = PolynomialFeatures(degree=2)
cubic = PolynomialFeatures(degree=3)
X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)
```

```python
# linear fit
X_fit = np.arange(X.min(), X.max(), 1)[:,np.newaxis]
regr = regr.fit(X,y)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y,regr.predict(X))
```

```python
# quadratic fit
regr = regr.fit(X_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quadratic_r2 = r2_score(y, regr.predict(X_quad))
```

```python
# cubic fit
regr = regr.fit(X_cubic, y)
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y, regr.predict(X_cubic))
```

```python
plt.scatter(X,y,label='training data', color='lightgray')
plt.plot(X_fit, y_lin_fit, 
         label='linear (d=1), $R^2=%.2f$' % linear_r2, 
         color='blue', 
         lw=2,
         linestyle=':')
plt.plot(X_fit, y_quad_fit, 
         label='quadratic (d=2), $R^2=%.2f$' % quadratic_r2, 
         color='red', 
         lw=2,
         linestyle='-')
plt.plot(X_fit, y_cubic_fit, 
         label='cubic (d=3), $R^2=%.2f$' % cubic_r2, 
         color='green', 
         lw=2,
         linestyle='--')
plt.xlabel('% lower status of population [LSTAT]')
plt.ylabel("Price in $1K [MEDV]")
plt.legend(loc='upper right')
```

```python
# transform features 
X_log = np.log(X)
```

```python
y_sqrt = np.sqrt(y)
```

```python
#fit
X_fit = np.arange(X_log.min()-1, X_log.max()+1,1)[:,np.newaxis]
regr = regr.fit(X_log, y_sqrt)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y_sqrt, regr.predict(X_log))
```

```python
plt.scatter(X_log, y_sqrt, label='training data', color='lightgray')
plt.plot(X_fit, y_lin_fit, 
         label='linear (d=1), $R^2=%.2f$' % linear_r2,
         color='blue', lw=2)
plt.xlabel('log(% lower status of population [LSTAT])')
plt.ylabel('$\sqrt{Price \; in \; \$1K [MEDV]}$')
plt.legend(loc='lower left')
```
